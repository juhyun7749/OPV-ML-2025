```python
import pandas



method_type = '2'   #
descriptors_df = pandas.read_csv('22.descriptors' + method_type + '.csv', index_col=[0, 1])   
operation_df = pandas.read_csv('00.operations.csv') 
operation_dict = dict(zip(operation_df['descriptor'], operation_df['operation'])) 
donor_df = pandas.read_csv('33.donor' + method_type + '.csv')  # 


descriptor_column_array = descriptors_df.columns[1:]

if method_type == '1':
    
    group_bb = pandas.DataFrame(columns = descriptor_column_array)
    group_sc = pandas.DataFrame(columns = descriptor_column_array)

    
    result_sheet_column_names_list = ['Donor']

    
    
    result_sheet_column_names_list += ['1_' + name for name in descriptor_column_array.to_list()]
    result_sheet_column_names_list += ['2_' + name for name in descriptor_column_array.to_list()]
    result_sheet_column_names_list += ['3_' + name for name in descriptor_column_array.to_list()]
    result_sheet_column_names_list += ['4_' + name for name in descriptor_column_array.to_list()]

    
    generated_df = pandas.DataFrame(columns=pandas.Index(result_sheet_column_names_list))

   
    group_bb.loc[0] = [0] * len(descriptor_column_array)
    group_sc.loc[0] = [0] * len(descriptor_column_array)

    
    for idx, row in descriptors_df.iterrows():
        if row.name[0] == 'BB':
            group_bb.loc[len(group_bb)] = row[1:].values
        else:
            group_sc.loc[len(group_sc)] = row[1:].values

    donor_column_index = [
        ['BB1', 'SC1'],
        ['BB2', 'SC2'],
        ['BB3', 'SC3'],
        ['BB4', 'SC4']
        ]

    
    for idx, row in donor_df.iterrows():
        if pandas.isna(row['Donor']):
            break
        
        
        a_descriptors = []
        b_descriptors = []

        new_row = [row['Donor']]

        
        ratioa = row['ratio']

        
        for col_idx, col in enumerate(donor_column_index):
            descriptor_dict = dict()
            
            bb_descript_value = group_bb.loc[row[donor_column_index[col_idx][0]]].copy()
            sc_descript_value = group_sc.loc[row[donor_column_index[col_idx][1]]].copy()

            
            num_value = 0
            sym_value = 0
            if col_idx == 0:
                num_value = row['num']
                sym_value = row['sym']
            else:
                num_value = row['num.' + str(col_idx)]
                sym_value = row['sym.' + str(col_idx)]

            for column_key in descriptor_column_array:
                if operation_dict[column_key] == 'max':
                    descriptor_dict[column_key] = max(bb_descript_value[column_key], sc_descript_value[column_key])

                elif operation_dict[column_key] == 'min':
                    descriptor_dict[column_key] = min(bb_descript_value[column_key], sc_descript_value[column_key])

                elif operation_dict[column_key] == 'avg':
                    if num_value > 0:
                        descriptor_dict[column_key] = (bb_descript_value[column_key] + sc_descript_value[column_key] * num_value) / (1 + num_value)
                    else:
                        descriptor_dict[column_key] = bb_descript_value[column_key]

                elif operation_dict[column_key] == 'sum':
                    if num_value > 0:
                        descriptor_dict[column_key] = bb_descript_value[column_key] + sc_descript_value[column_key] * num_value
                    else:
                        descriptor_dict[column_key] = bb_descript_value[column_key]

            a_descriptors += list(descriptor_dict.values())

          
        ratiob = row['ratio.1']

        
        for col_idx, col in enumerate(donor_column_index):
            descriptor_dict = dict()
            
            bb_descript_value = group_bb.loc[row[donor_column_index[col_idx][0] + ".1"]].copy()
            sc_descript_value = group_sc.loc[row[donor_column_index[col_idx][1] + ".1"]].copy()

            num_value = row['num.' + str(col_idx + 3)]
            sym_value = row['sym.' + str(col_idx + 3)]

            for column_key in descriptor_column_array:
                if operation_dict[column_key] == 'max':
                    descriptor_dict[column_key] = max(bb_descript_value[column_key], sc_descript_value[column_key])

                elif operation_dict[column_key] == 'min':
                    descriptor_dict[column_key] = min(bb_descript_value[column_key], sc_descript_value[column_key])

                elif operation_dict[column_key] == 'avg':
                    if num_value > 0:
                        descriptor_dict[column_key] = (bb_descript_value[column_key] + sc_descript_value[column_key] * num_value) / (1 + num_value)
                    else:
                        descriptor_dict[column_key] = bb_descript_value[column_key]

                elif operation_dict[column_key] == 'sum':
                    if num_value > 0:
                        descriptor_dict[column_key] = bb_descript_value[column_key] + sc_descript_value[column_key] * num_value
                    else:
                        descriptor_dict[column_key] = bb_descript_value[column_key]
            
            b_descriptors += list(descriptor_dict.values())
            
       
        new_row += [a * ratioa + b * ratiob for a, b in zip(a_descriptors, b_descriptors)]
        generated_df.loc[len(generated_df)] = new_row

    generated_df = generated_df.loc[:, (generated_df != 0).any(axis=0)]
    generated_df.to_csv("44.donor_result1.csv", index = False)
else:
    
    part_1 = pandas.DataFrame(columns = descriptor_column_array)
    part_2 = pandas.DataFrame(columns = descriptor_column_array)
    part_3 = pandas.DataFrame(columns = descriptor_column_array)
    part_4 = pandas.DataFrame(columns = descriptor_column_array)

   
    result_sheet_column_names_list = ['Donor']

   
    
    result_sheet_column_names_list += ['1_' + name for name in descriptor_column_array.to_list()]
    result_sheet_column_names_list += ['2_' + name for name in descriptor_column_array.to_list()]
    result_sheet_column_names_list += ['3_' + name for name in descriptor_column_array.to_list()]
    result_sheet_column_names_list += ['4_' + name for name in descriptor_column_array.to_list()]

    
    
    generated_df = pandas.DataFrame(columns = pandas.Index(result_sheet_column_names_list))

    part_1.loc[0] = [0] * len(descriptor_column_array)
    part_2.loc[0] = [0] * len(descriptor_column_array)
    part_3.loc[0] = [0] * len(descriptor_column_array)
    part_4.loc[0] = [0] * len(descriptor_column_array)

    for idx, row in descriptors_df.iterrows():
        if row.name[0] == 'Part 1':
            part_1.loc[len(part_1)] = row[1:].values
        elif row.name[0] == 'Part 2':
            part_2.loc[len(part_2)] = row[1:].values
        elif row.name[0] == 'Part 3':
            part_3.loc[len(part_3)] = row[1:].values
        else:
            part_4.loc[len(part_4)] = row[1:].values

    for idx, row in donor_df.iterrows():
      
        new_row = [row['Donor']]
        
        ratioa = row['ratio']
        ratiob = row['ratio.1']

        

        new_row += [a * ratioa + b * ratiob for a, b in zip(part_1.loc[row['Scaffold1']].to_list(), part_1.loc[row['Scaffold1.1']].to_list())]
        new_row += [a * ratioa + b * ratiob for a, b in zip(part_2.loc[row['Scaffold2']].to_list(), part_2.loc[row['Scaffold2.1']].to_list())]
        new_row += [a * ratioa + b * ratiob for a, b in zip(part_3.loc[row['Scaffold3']].to_list(), part_3.loc[row['Scaffold3.1']].to_list())]
        new_row += [a * ratioa + b * ratiob for a, b in zip(part_4.loc[row['Scaffold4']].to_list(), part_4.loc[row['Scaffold4.1']].to_list())]

        generated_df.loc[len(generated_df)] = new_row

    generated_df = generated_df.loc[:, (generated_df != 0).any(axis=0)]
    generated_df.to_csv("44.donor_result2.csv", index = False)

```
